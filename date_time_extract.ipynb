**working with date**

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

```python
# extract the month
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
# extract the month_name
date['date_month_name']=date['date'].dt.month_name()
date.head()
```

```python
#extract the day
date['date_day']=date['date'].dt.day
date.head()
```

```python
# extract the day of week
date['date_dow']=date['date'].dt.dayofweek
date.head()
```

```python
# extract day nam4
date['date_dow_name']=date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# isin weekeend
date['date_is_weekend']=np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
# extract the week of year
# extract the week of year
date['date_week']=date['date'].dt.isocalendar().week
date.head()
```

```python
# extract the quarter
date['quarter']=date['date'].dt.quarter
date.head()
```

```python
#extract semeester
date['semester']=np.where(date['quarter'].isin([1,2]),1,2)
date.head()
```

```python
#time
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date'] # extract time elapse
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

# Calculate the difference in days
days_passed = (today - date['date']).dt.days

# Approximate the number of months passed by dividing by the average number of days in a month
# Using 30.44 as an approximation for the average number of days in a month (365.25 / 12)
months_passed = np.round(days_passed / 30.44, 0)

# You can assign this to a new column if needed
# date['months_passed'] = months_passed

# Display the calculated months passed (optional)
print(months_passed.head())
```
